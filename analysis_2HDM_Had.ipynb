```python
def signal_label(param_0, param_1):
    label = r'$m_H=$' + str(param_0) + r', $m_\mathit{a}=$' + str(param_1)
    return label

def plot_analysis( dataframes, labels, colors, title1,
                    var, xlabel1, binMin1,binMax1,binn1, ylim_max,
                    # folder, 
                    binagem=None , cut_applied=None, above=True ):
    
    fig1 = plt.figure(figsize=(8,7))
    grid = [2, 1] #Número de colunas
    gs1 = gs.GridSpec(grid[0], grid[1], height_ratios=[4, 1])

    #=================================================================================================================
    N = 1 
    #=================================================================================================================
    #==================================================
    ax1 = plt.subplot(ana.position(gs1,grid,N,1))              # Positioning at subplot 1 of the plot number 1
    #==================================================
    # var = "Jet_mass"
    bins = np.linspace(binMin1,binMax1,binn1)
    if binagem != None:
        bins = binagem
    ana.step_plot( ax1, var, df['Signal_400_100'], label=signal_label(400,100), color='blue', weight="evtWeight", bins=bins )
    ana.step_plot( ax1, var, df['Signal_1000_100'], label=signal_label(1000,100), color='turquoise', weight="evtWeight", bins=bins )
    # ana.step_plot( ax1, var, df_400_100, label=signal_label(400,100), color='slategray', weight="evtWeight", bins=bins )
    ybkg, errbkg = ana.stacked_plot( ax1, var, dataframes, labels, colors, weight="evtWeight", bins=bins )  # Produce the stacked plot
    # ydata, errdata = ana.data_plot( ax1, var, df_DATA, bins=bins )
    ana.labels(ax1, ylabel="Events")  # Set up the label names
    # ana.style(ax1, lumi=35.9, year=2016, ylog=True, legend_ncol=2, ylim=[1.e-1,1.e6], xticklabels=False) # Set up the plot style and information on top
    ana.style(ax1, lumi=35.9, year=2022, ylog=True, legend_ncol=2, ylim=[1.e-1,ylim_max], xticklabels=False)
    ax1.set_title(f"{title1}\n\n", {'fontsize': 15, 'color': 'red'})
    # For vertical lines
    if cut_applied is not None:
        i=0
        for cut in cut_applied:
            x = cut
            y = 0.5+i*0.05
            plt.axvline(x=x, color='grey', linestyle='--') # vertical lines
            # plt.text(x+0.005, y, str(cut))
            if above==True:
                plt.text(x + 0.005, y, fr'{cut} $\rightarrow$')
            else:
                plt.text(x + 0.005, y, fr'{cut} $\leftarrow$')
            i = i+1
    #==================================================
    ax2 = plt.subplot(ana.position(gs1,grid,N,2), sharex=ax1)  # Positioning at subplot 2 of the plot number 2
    #==================================================
    # ctr = ana.control( var, [df['Signal_400_100']], dataframes, weight="evtWeight", bins=np.linspace(0.8,3.2,1001) ) bins
    ctr = ana.control( var, [df['Signal_400_100']], dataframes, weight="evtWeight", bins=bins )
    #ctr.purity_plot()
    ctr.signal_eff_plot(label='Signal_400_100 efficiency')
    ctr.bkg_eff_plot()
    #-------------------------
    # For vertical lines
    if cut_applied is not None:
        i=0
        for cut in cut_applied:
            x = cut
            y = 0.5+i*0.05
            plt.axvline(x=x, color='grey', linestyle='--') # vertical lines
            # plt.text(x+0.005, y, str(cut))
            i = i+1
    ana.labels(ax2, xlabel=xlabel1, ylabel="Control")  # Set up the label names
    ana.style(ax2, ylim=[0., 1.1], yticks=[0., 0.2, 0.4, 0.6, 0.8, 1.], xgrid=True, ygrid=True)


def plot_roc_analysis( dataframes, labels, colors, title1,
                    var, xlabel1, binMin1,binMax1,binn1, ylim_max,
                    folder, binagem=None, cut_applied=None, above=True ):
    
    #==================================================================================================
    # Plot
    #==================================================================================================
    # fig1 = plt.figure(figsize=(20,7.5))
    # grid = [2, 3] #Número de colunas
    # gs1 = gs.GridSpec(grid[0], grid[1], height_ratios=[4, 1])
    fig1 = plt.figure(figsize=(8,7))
    grid = [2, 1] #Número de colunas
    gs1 = gs.GridSpec(grid[0], grid[1], height_ratios=[4, 1])

    #=================================================================================================================
    N = 1
    #=================================================================================================================
    #==================================================
    ax1 = plt.subplot(ana.position(gs1,grid,N,1))              # Positioning at subplot 1 of the plot number 1
    #==================================================
    # var = "MET_pt"
    bins = np.linspace(binMin1,binMax1,binn1)
    if binagem != None:
        bins = binagem
    # ana.step_plot( ax1, var, dataframes, labels, color='black', weight="evtWeight", bins=bins )
    #-------------------------
    ctr = ana.control( var, [df["Signal_400_100"]], dataframes, weight="evtWeight", bins=np.linspace(binMin1,binMax1,200), above=above )
    ctr.roc_plot(label='Signal_400_100', color='blue', linestyle="-")
    #-------------------------
    ctr = ana.control( var, [df["Signal_1000_100"]], dataframes, weight="evtWeight", bins=np.linspace(binMin1,binMax1,200), above=above )
    ctr.roc_plot(label='Signal_1000_100', color='turquoise', linestyle="-")
    #-------------------------
    # ctr = ana.control( var, [df["Signal_1000_800"]], dataframes, weight="evtWeight", bins=np.linspace(binMin1,binMax1,200), above=above)
    # ctr.roc_plot(label='Signal_1000_800', color='turquoise', linestyle="-")
    #-------------------------
    # For vertical lines
    if cut_applied is not None:
        i=0
        for cut in cut_applied:
            x = 1 - ctr.bkg_eff( cut, apx=True)
            # if below==True:
            y = 0.5+i*0.05
            plt.axvline( x=x, color='grey', linestyle='--') # vertical lines
            plt.text(x+0.005, y, str(cut))
            
            # plt.axhline(1, color='grey', linewidth=1, linestyle="dotted") # linha horizontal
            i = i+1  
    # ana.labels(ax1, xlabel=xlabel1+' [Background Rejection]', ylabel="Signal Efficiency")  # Set up the label names
    ana.labels(ax1, 'Background Rejection', ylabel="Signal Efficiency")  # Set up the label names
    ana.style(ax1, xgrid=True, ygrid=True, legend_loc='lower left')
```
